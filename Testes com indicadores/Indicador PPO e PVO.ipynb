```python
from stock_indicators import indicators
from stock_indicators import Quote
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
```

```python
ticker ="CRFB3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2022":]
```

```python
resultado_ppo = ta.momentum.PercentagePriceOscillator(df["Close"],26,12,9,False)


ppo_list = list()
ppo_hist = list()
ppo_signal = list()

i=0
while i< len(resultado_ppo.ppo()):
    
    if(resultado_ppo.ppo()[i] != None):
        
        ppo_list.append(resultado_ppo.ppo()[i])
        
    i+=1
    
    
i=0
while i< len(resultado_ppo.ppo_hist()):
    
    if(resultado_ppo.ppo_hist()[i] != None):
        
        ppo_hist.append(resultado_ppo.ppo_hist()[i])
        
    i+=1
    
i=0
while i< len(resultado_ppo.ppo_signal()):
    
    if(resultado_ppo.ppo_signal()[i] != None):
        
        ppo_signal.append(resultado_ppo.ppo_signal()[i])
        
    i+=1
    
    
fig = go.Figure()

fig.add_trace(
    go.Bar(
    name="Histograma",
    x=df.index,
    y=ppo_hist,
    marker_color='blue'
))

fig.add_trace(
    go.Scatter(
        name="Sinal",
        x=df.index,
        y=ppo_signal,
        line=dict(color='blue', width=1))
)

fig.add_trace(
    go.Scatter(
        name="Sinal",
        x=df.index,
        y=ppo_list,
        line=dict(color='red', width=1))
)

fig.update_layout(xaxis_rangeslider_visible=True)
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```
