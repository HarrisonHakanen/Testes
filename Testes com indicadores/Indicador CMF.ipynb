```python
from stock_indicators import indicators
from stock_indicators import Quote
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
```

```python
ticker ="CRFB3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2020":]
```

```python
resultado_cmf = ta.volume.ChaikinMoneyFlowIndicator(df["High"],df["Low"],df["Close"],df["Volume"],20,False)
```

```python
cmf_list = list()


i=0
while i< len(resultado_cmf.chaikin_money_flow()):
    
    if(resultado_cmf.chaikin_money_flow()[i] != None):
        
        cmf_list.append(resultado_cmf.chaikin_money_flow()[i])
        
    i+=1
    
```

```python
fig = go.Figure()



fig.add_trace(
    go.Bar(
    x=df.index,
    y=cmf_list,
    marker_color='blue'
))



fig.add_hline(y=0,line_color="blue",line_width=1)
#fig.update_layout(xaxis_rangeslider_visible=True)
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="Fechamento",
        x=df.index,
        y=df["Close"],
        line=dict(color='blue', width=2))
)
```
