```python
#IMPLEMENTAR O KAMA E O TSI


from stock_indicators import indicators
from stock_indicators import Quote
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
```

```python
ticker ="PETR3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2022":]
```

```python
resultados_kama = ta.momentum.KAMAIndicator(df["Close"],10,2,30,False)

kama_list = list()

i=0
while i< len(resultados_kama.kama()):
    
    if(resultados_kama.kama()[i] != None):
        
        kama_list.append(resultados_kama.kama()[i])
        
    i+=1
    
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="KAMA",
        x=df.index,
        y=kama_list,
        line=dict(color='blue', width=1))
)

fig.add_trace(go.Candlestick(
        x=df.index,
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close']))

```

```python
resultados_roc = ta.momentum.ROCIndicator(df["Close"],12,False)

roc_list = list()

i=0
while i< len(resultados_roc.roc()):
    
    if(resultados_roc.roc()[i] != None):
        
        roc_list.append(resultados_roc.roc()[i])
        
    i+=1
    
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="ROC",
        x=df.index,
        y=roc_list,
        line=dict(color='blue', width=1))
)

fig.add_trace(go.Candlestick(
        x=df.index,
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close']))

```

```python
resultados_rsi = ta.momentum.StochRSIIndicator(df["Close"],14,3,3,False)

stochrsi_list = list()
d_list = list()
k_list = list()


i=0
while i< len(resultados_rsi.stochrsi()):
    
    if(resultados_rsi.stochrsi()[i] != None):
        
        stochrsi_list.append(resultados_rsi.stochrsi()[i])
        
    i+=1
    
    
i=0
while i< len(resultados_rsi.stochrsi_d()):
    
    if(resultados_rsi.stochrsi_d()[i] != None):
        
        d_list.append(resultados_rsi.stochrsi_d()[i])
        
    i+=1


i=0
while i< len(resultados_rsi.stochrsi_k()):
    
    if(resultados_rsi.stochrsi_k()[i] != None):
        
        k_list.append(resultados_rsi.stochrsi_k()[i])
        
    i+=1
    
    
    
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="Stochrsi",
        x=df.index,
        y=stochrsi_list,
        line=dict(color='blue', width=1))
)

fig.add_trace(
    go.Scatter(
        name="%d",
        x=df.index,
        y=d_list,
        line=dict(color='red', width=1))
)

fig.add_trace(
    go.Scatter(
        name="%k",
        x=df.index,
        y=k_list,
        line=dict(color='green', width=1))
)
```

```python
resultados_stochastic = ta.momentum.StochasticOscillator(df["High"], df["Low"], df["Close"], 14, 3, False)

stochastic_list = list()
signal_list = list()

i=0
while i< len(resultados_stochastic.stoch()):
    
    if(resultados_stochastic.stoch()[i] != None):
        
        stochastic_list.append(resultados_stochastic.stoch()[i])
        
    i+=1
    
    
i=0
while i< len(resultados_stochastic.stoch_signal()):
    
    if(resultados_stochastic.stoch_signal()[i] != None):
        
        signal_list.append(resultados_stochastic.stoch_signal()[i])
        
    i+=1
    
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="Stochrsi",
        x=df.index,
        y=stochastic_list,
        line=dict(color='blue', width=1))
)

fig.add_trace(
    go.Scatter(
        name="Sinal",
        x=df.index,
        y=signal_list,
        line=dict(color='red', width=1))
)
```

```python
resultados_tsi = ta.momentum.TSIIndicator(df["Close"],25,13,False)

tsi_list = list()

i=0
while i< len(resultados_tsi.tsi()):
    
    if(resultados_tsi.tsi()[i] != None):
        
        tsi_list.append(resultados_tsi.tsi()[i])
        
    i+=1
    
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="TSI",
        x=df.index,
        y=tsi_list,
        line=dict(color='blue', width=1))
)

fig.add_hline(y=20,line_color="blue",line_width=1, line_dash="dash")
fig.add_hline(y=-20,line_color="blue",line_width=1, line_dash="dash")
```
