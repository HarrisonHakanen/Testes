```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from   sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from   sklearn.metrics import r2_score
import statsmodels.api as sm
import os

import yfinance as yf
from datetime import datetime

import plotly
import plotly.express as px
import pandas as pd

import dateutil.relativedelta
from datetime import datetime

import plotly.graph_objects as go
```

```python
ticker ="SOMA3.SA"

acao = yf.download(ticker)

acao_regressao = acao["2020-04-30":"2022-06-14"]
acao_regressao.reset_index(inplace=True)

X = acao_regressao.index.values.reshape(-1,1)
y = acao_regressao['Close'].values.reshape(-1,1)

#Regressão linear
reg = LinearRegression()
reg.fit(X, y)

f_previsoes = reg.predict(X)

previsoes_lista = list()
i = 0

while i < len(f_previsoes):
    
    previsoes_lista.append(f_previsoes[i][0])
    
    i+=1

#Regressão polinomial
poly = PolynomialFeatures(degree = 4)
X_poly = poly.fit_transform(X)
  
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)

f_polinomial = lin2.predict(poly.fit_transform(X))

previsoes_polinomial = list()

i = 0

while i < len(f_polinomial):
    
    previsoes_polinomial.append(f_polinomial[i][0])
    
    i+=1



fig = go.Figure()

fig.add_trace(go.Candlestick(
name='Fechamento', 
x=acao_regressao["Date"], 
open=acao_regressao["Open"], 
high=acao_regressao["High"],
low=acao_regressao["Low"],
close=acao_regressao["Close"]))

fig.add_trace(
    go.Scatter(
        name="Regressão linear",
        x=acao_regressao["Date"],
        y=previsoes_lista,
        line=dict(color='red', width=1))
)

fig.add_trace(
    go.Scatter(
        name="Regressão polinomial",
        x=acao_regressao["Date"],
        y=previsoes_polinomial,
        line=dict(color='blue', width=1))
)



fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')

fig.show()
```
