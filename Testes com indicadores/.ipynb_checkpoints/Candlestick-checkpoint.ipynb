```python
import yfinance as yf
from datetime import datetime

import plotly
import plotly.express as px
import pandas as pd

import dateutil.relativedelta
from datetime import datetime

import plotly.graph_objects as go
```

```python
ticker ="AMER3.SA"

acao = yf.download(ticker)

data_final = datetime.now()

data_inicial = data_final + dateutil.relativedelta.relativedelta(months=-4)

acao = acao[data_inicial:data_final]

acao['MM10'] = acao["Close"].rolling(10).mean()
acao['MM20'] = acao["Close"].rolling(20).mean()
acao['MM30'] = acao["Close"].rolling(30).mean()
acao['MM05'] = acao["Close"].rolling(5).mean()
    
#Técnica borboleta---------------------------------------------------------


fig = go.Figure()

fig.add_trace(go.Candlestick(
    name='Fechamento', 
    x=acao.index, 
    open=acao["Open"], 
    high=acao["High"],
    low=acao["Low"],
    close=acao["Close"],
    increasing_line_color= '#B2DEFF', decreasing_line_color= '#194362'))

fig.add_trace(
    go.Scatter(
        name="MM 30",
        x=acao.index,
        y=acao["MM30"],
        line=dict(color='#FFC300', width=1, dash='dot'))
)
fig.add_trace(
    go.Scatter(
        name="MM 20",
        x=acao.index,
        y=acao["MM20"],
        line=dict(color='#FF5733', width=1, dash='dot'))
)
fig.add_trace(
    go.Scatter(
        name="MM 10",
        x=acao.index,
        y=acao["MM10"],
        line=dict(color='#900C3F', width=1, dash='dot'))
)

fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')

fig.show()



#---------------------------------------------------------------------------------------

fig2 = go.Figure()

fig2.add_trace(go.Candlestick(
    name='Fechamento', 
    x=acao.index, 
    open=acao["Open"], 
    high=acao["High"],
    low=acao["Low"],
    close=acao["Close"],
    increasing_line_color= '#B2DEFF', decreasing_line_color= '#194362'))

fig2.add_trace(
    go.Scatter(
        name="MM 5",
        x=acao.index,
        y=acao["MM05"],
        line=dict(color='#FFC300', width=1, dash='dot')))

fig2.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')

fig2.show()



#---------------------------------------------------------------------------------------
fig3 = go.Figure()

fig3.add_trace(go.Candlestick(
    name='Fechamento', 
    x=acao.index, 
    open=acao["Open"], 
    high=acao["High"],
    low=acao["Low"],
    close=acao["Close"]))

fig3.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')

fig3.show()
```
