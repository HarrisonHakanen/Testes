```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go

df = yf.download("PETR3.SA", interval = "1m",period="1d")
df.reset_index(inplace=True)
```

```python
df= df[0:50]
```

```python
df
```

```python
fig = go.Figure()
fig.add_trace(go.Candlestick(
    name='Fechamento', 
    x=df["Datetime"], 
    open=df["Open"], 
    high=df["High"],
    low=df["Low"],
    close=df["Close"]))
```
