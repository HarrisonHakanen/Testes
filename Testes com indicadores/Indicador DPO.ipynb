```python
from stock_indicators import indicators
from stock_indicators import Quote
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
```

```python
ticker ="PETR3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2022":]
```

```python
dpo_resultado = ta.trend.DPOIndicator(df["Close"],20,False)
```

```python
dpo_list = list()

i=0
while i< len(dpo_resultado.dpo()):
    
    if(dpo_resultado.dpo()[i] != None):
        
        dpo_list.append(dpo_resultado.dpo()[i])
        
    i+=1
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="DPO",
        x=df.index,
        y=dpo_list,
        line=dict(color='blue', width=1))
)

fig.add_hline(y=1,line_color="blue",line_width=1, line_dash="dash")
fig.add_hline(y=-1,line_color="blue",line_width=1, line_dash="dash")
fig.update_layout(xaxis_rangeslider_visible=True)
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```
