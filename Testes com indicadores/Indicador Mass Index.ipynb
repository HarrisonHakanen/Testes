```python
from stock_indicators import indicators
from stock_indicators import Quote
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
```

```python
ticker ="PETR3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2022":]
```

```python
resultados_mi = ta.trend.MassIndex(df["High"], df["Low"],9,25,False)
```

```python
mi_list = list()


i=0
while i< len(resultados_mi.mass_index()):
    
    if(resultados_mi.mass_index()[i] != None):
        
        mi_list.append(resultados_mi.mass_index()[i])
        
    i+=1
    
mi_df = pd.DataFrame(mi_list,columns=["mi"])
media_mi = mi_df["mi"].rolling(9).mean()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="Mass Index",
        x=df.index,
        y=mi_list,
        line=dict(color='blue', width=1))
)
fig.add_trace(
    go.Scatter(
        name="Media",
        x=df.index,
        y=media_mi,
        line=dict(color='green', width=1))
)

fig.update_layout(xaxis_rangeslider_visible=True)
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```
