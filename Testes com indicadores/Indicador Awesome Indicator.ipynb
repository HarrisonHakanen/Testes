```python
from stock_indicators import indicators
from stock_indicators import Quote
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
```

```python
ticker ="PETR3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2022":]
```

```python
resultados_awesome = ta.momentum.AwesomeOscillatorIndicator(df["High"],df["Low"],5,34,False)
```

```python
awesome_list = list()


i=0
while i< len(resultados_awesome.awesome_oscillator()):
    
    if(resultados_awesome.awesome_oscillator()[i] != None):
        
        awesome_list.append(resultados_awesome.awesome_oscillator()[i])
        
    i+=1
    
```

```python
fig = go.Figure()

fig.add_trace(
    go.Bar(
    x=df.index,
    y=awesome_list,
    marker_color='blue'
))


fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="Fechamento",
        x=df.index,
        y=df["Close"],
        line=dict(color='blue', width=2))
)
```
