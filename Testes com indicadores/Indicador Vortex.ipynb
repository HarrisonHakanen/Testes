```python
from stock_indicators import indicators
from stock_indicators import Quote
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
```

```python
ticker ="PETR3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2022":]
```

```python
resultados_vortex = ta.trend.VortexIndicator(df["High"],df["Low"],df["Close"],14,False)
```

```python
vortex_neg = list()
vortex_pos = list()

i=0
while i< len(resultados_vortex.vortex_indicator_neg()):
    
    if(resultados_vortex.vortex_indicator_neg()[i] != None):
        
        vortex_neg.append(resultados_vortex.vortex_indicator_neg()[i])
        
    i+=1
    
    

i=0
while i< len(resultados_vortex.vortex_indicator_pos()):
    
    if(resultados_vortex.vortex_indicator_pos()[i] != None):
        
        vortex_pos.append(resultados_vortex.vortex_indicator_pos()[i])
        
    i+=1
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="Positivo",
        x=df.index,
        y=vortex_pos,
        line=dict(color='blue', width=1))
)

fig.add_trace(
    go.Scatter(
        name="Negativo",
        x=df.index,
        y=vortex_neg,
        line=dict(color='red', width=1))
)

fig.add_hline(y=1,line_color="blue",line_width=1, line_dash="dash")
fig.update_layout(xaxis_rangeslider_visible=True)
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```
