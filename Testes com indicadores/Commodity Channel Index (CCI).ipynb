```python
from stock_indicators import indicators
from stock_indicators import Quote
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
```

```python
ticker ="PETR3.SA"

#df = yf.download(ticker,period="1y", interval="90m")
df = yf.download(ticker)
df=df["2022":]
df2=df["2022":]

#Prepara os dados

df2.reset_index(inplace=True)

df2.rename(columns={"Date":"date","Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume"},inplace=True)

quotes_list = [
    Quote(d,o,h,l,c,v) 
    for d,o,h,l,c,v 
    in zip(df2['date'], df2['open'], df2['high'], df2['low'], df2['close'], df2['volume'])
]
```

```python
CCI_Results = indicators.get_cci(quotes_list, 20)

#Separa os dados
i=0
CCI_List = list()
CCI_Date = list()

while i< len(CCI_Results):
    
    
    if(CCI_Results[i].cci != None):
        
        CCI_List.append(CCI_Results[i].cci)
        
    i+=1
```

```python
CCI_TA = ta.trend.CCIIndicator(df["High"],df["Low"],df["Close"],20,0.015,False)
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="CCI",
        x=CCI_TA.cci().index,
        y=CCI_TA.cci(),
        line=dict(color='blue', width=2))
)

fig.add_hline(y=-100,line_color="blue",line_width=1, line_dash="dash")
fig.add_hline(y=100,line_color="blue",line_width=1, line_dash="dash")

fig.update_layout(xaxis_rangeslider_visible=True)
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```
