```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import ta as ta
from ta import add_all_ta_features
from ta.utils import dropna
from datetime import datetime
from tsmoothie.utils_func import sim_randomwalk
from tsmoothie.smoother import *
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go



import statsmodels.api as sm
```

```python
ticker ="PETR3.SA"

df = yf.download(ticker)

base = df["2018":]


#TSI

tsi_config=[25,13]
resultados_tsi = ta.momentum.TSIIndicator(base["Close"],tsi_config[0],tsi_config[1],False)

resultados_tsi.tsi().dropna(inplace=True)


#Gaussian TSI
resultados_tsi_sem_null = resultados_tsi.tsi().dropna()

tsi_np = resultados_tsi_sem_null.to_numpy()

tsi_np = tsi_np.reshape(1,len(tsi_np))


smoother_tsi = GaussianSmoother(n_knots=60, sigma=0.0003)
smoother_tsi.smooth(tsi_np)

tsi_suavizado = smoother_tsi.smooth_data[0]

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name="TSI",
        x=base.tail(len(resultados_tsi_sem_null)).index,
        y=resultados_tsi_sem_null,
        line=dict(color='blue', width=1))
)

fig.add_trace(
    go.Scatter(
        x=base.tail(len(resultados_tsi_sem_null)).index,
        y=tsi_suavizado,
        mode='lines',
        name='Gaussian',
        line=dict(color="orange"))
)


fig.add_hline(y=0,line_color="blue",line_width=1, line_dash="dash")
fig.update_layout(paper_bgcolor='rgba(0,0,0,0)', plot_bgcolor='rgba(0,0,0,0)')
```

```python
tsi_suavizado_df = pd.DataFrame(tsi_suavizado,columns=["tsi_suavizado"])
```

```python
ticker ="PETR3.SA"

df = yf.download(ticker)

base = df["2018":]

# Calcular estatísticas de retorno
returns = tsi_suavizado_df["tsi_suavizado"].pct_change().dropna()
mean_return = returns.mean()
volatility = returns.std()

# Gerar cenários futuros
num_simulations = 30
num_days = 2

simulated_returns = np.random.normal(mean_return, volatility, (num_simulations, num_days))

# Simular preços futuros
last_price = tsi_suavizado_df["tsi_suavizado"].iloc[-1]
simulated_prices = np.zeros((num_simulations, num_days + 1))
simulated_prices[:, 0] = last_price
for i in range(1, num_days + 1):
    simulated_prices[:, i] = simulated_prices[:, i - 1] * (1 + simulated_returns[:, i - 1])

# Analisar resultados
plt.plot(simulated_prices)
plt.xlabel("Dias")
plt.ylabel("Preço")
plt.title("Simulação de Monte Carlo para previsão de preços de ativos")
plt.show()
```
